```python
import pandas as pd
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
print (df.describe())
```

```python
X = df.drop("Outcome" , axis= 1)
y= df['Outcome']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X_train, X_test , y_train , y_test = train_test_split(X , y, test_size = 0.3, random_state= 40)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('Classifier', RandomForestClassifier())
])
```

```python
param_grid = {
    'Classifier__n_estimators': [50, 100, 200],  # Hyperparameter for RandomForestClassifier
    'Classifier__max_depth': [None, 10, 20, 30],  # Hyperparameter for RandomForestClassifier
    'Classifier__min_samples_split': [2, 5, 10],  # Hyperparameter for RandomForestClassifier
}
```

```python
grid_search = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=-1, verbose=1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
best_params = grid_search.best_params_
best_params
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model
```

```python
y_pred = best_model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Best Hyperparameters: {grid_search.best_params_}")
print(f"Test Accuracy: {accuracy:.4f}")
```
