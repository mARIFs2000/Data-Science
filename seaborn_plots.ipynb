```python
import seaborn as sns
import matplotlib.pyplot as plt
penguin = sns.load_dataset('penguins')
penguin
```

```python
sns.scatterplot(x="island",y="bill_length_mm",data = penguin)
plt.show()
```

```python
sns.violinplot(x="bill_length_mm",y="island",data = penguin)
plt.show()
```

```python
sns.boxplot(x="island",y="bill_length_mm",data = penguin)
plt.show()
```

```python
sns.pairplot(penguin,hue='species')
plt.show()
```

```python
sns.displot(penguin['body_mass_g'],kind= 'kde')
plt.show()
```

```python
sns.displot(penguin,x='flipper_length_mm',hue='sex',kind= 'hist',bins=25)
plt.show()
```
