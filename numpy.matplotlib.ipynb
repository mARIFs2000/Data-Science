```python
import matplotlib.pyplot as plt
import numpy as np
y1 = np.array ([9,0,6,4,8,2])
x1 = np.array ([2,8,4,6,0,9])
x2 = np.array ([1,9,3,6,8,0])
y2 = np.array ([0,8,6,3,9,1])
font1 =  {'family':'Garamond','color':'red','size':15}
font2 =  {'family':'serif', 'color':'green', 'size':18}

plt.title('Gaming Levels',fontdict = font2)
plt.xlabel('Speed',fontdict = font1)
plt.ylabel('RPM',fontdict = font1)

plt.plot(x1,y1,x2,y2)
plt.show()
```
